```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def f(x):
     return np.where(x<0,x*np.sin(x**2),np.cos(x))
x_points =np.linspace(-5,5,num=1000001)
print(x_points)

function_values = f(x_points)
print("All calculation at x_points follows:",function_values)
plt.plot(x_points,function_values,label='f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('plot of f(x)')

```
