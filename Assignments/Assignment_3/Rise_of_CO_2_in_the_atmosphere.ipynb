```python
import urllib.request

url = "ftp://aftp.cmdl.noaa.gov/products/trends/co2/co2_mm_mlo.txt"
data_file = "co2_mm_mlo.txt"

urllib.request.urlretrieve(url, data_file)
```

```python
import numpy as np

data = np.genfromtxt(data_file, skip_header=72)
```

```python
co2_monthly_averages = np.ma.masked_invalid(data[:, 3])
```

```python
import matplotlib.pyplot as plt

time = data[:, 0] + data[:, 1] / 12

plt.figure(figsize=(10, 6))
plt.plot(time, co2_monthly_averages, marker='o', linestyle='-', color='b')
plt.xlabel("Time")
plt.ylabel("CO2 Monthly Average")
plt.title("Monthly Averages of Measured CO2")
plt.grid(True)
plt.show()
```

```python
co2_rate_of_change = np.diff(co2_monthly_averages)
time_rate_of_change = time[1:]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time_rate_of_change, co2_rate_of_change, marker='o', linestyle='-', color='g')
plt.xlabel("Time")
plt.ylabel("Rate of Change of CO2")
plt.title("Estimate of Rate of Change of CO2")
plt.grid(True)
plt.show()
```
