```python
import numpy as np

n = 1000  #  Number of grid points
h = 2 * np.pi / (n - 1)  # Spacing of Grid

a=np.diag(np.ones(n - 1),k=1)
b=np.diag(-np.ones(n ),k=0)

#  The differentiation matrix D
D = np.delete(a+b,n-1, 0)
x = np.linspace(0, 2*np.pi, n)

f_x=np.sin(x)
df = np.dot(D, f_x/h)

#Actual derivative of sin(x)
y=np.cos(x)

# Ploting the results
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(x[:-1], df,'--', label="Approximate. Derivative",color='blue')
plt.plot(x, y,':',label='cos(x)', color='red')
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Approximate Derivative of f(x) =sin(x)")
plt.grid(False)
plt.legend()
plt.show()
```

```python
f_new = x**2 + 2 * np.sin(10 * x)

df_new = np.dot(D, f_new/h)

y1=2*x+20*np.cos(10*x)

plt.figure(figsize=(10, 6))
plt.plot(x[:-1], df_new, label="Approx. Derivative")
plt.plot(x, y1,':',label='2x+20cos(10x)', color='red')
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Approximate Derivative of f(x) = x^2 + 2sin(10x)")
plt.grid(False)
plt.legend()
plt.show()
```

```python
from scipy.sparse import diags

n = 1000  # Number of grid points n
h = 2 * np.pi / n  # Grid spacing h
main_diagonal = np.ones(n - 1)
off_diagonal = -np.ones(n - 1)

D_sparse = diags([main_diagonal, off_diagonal], [1, 0], shape=(n - 1, n))
x = np.linspace(0, 2 * np.pi, n)

f_x1 = np.sin(x)
f_x2 = x**2 + 2 * np.sin(10 * x)

df1_sparse = D_sparse.dot(f_x1) / h
df2_sparse = D_sparse.dot(f_x2) / h

plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
plt.plot(x[:-1], df1_sparse, label="Sparse Approximate. Derivative (sin(x))", color='blue')
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Sparse Approximate Derivative of f(x) = sin(x)")
plt.grid(False)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x[:-1], df2_sparse, label="Sparse Approximate. Derivative (x^2 + 2*sin(10x))", color='green')
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Sparse Approximate Derivative of f(x) = x^2 + 2*sin(10x)")
plt.grid(False)
plt.legend()

plt.tight_layout()
plt.show()
```
